```python
# Get images with one specific representation (among others)

study = api.BIAStudy.get_by_accession("S-BIAD704")

ome_ngff_images = list(study.get_images_with_representation(api.ImageRepresentationType.OME_NGFF))
print(len(ome_ngff_images))
example_image = next(study.get_images_with_representation(api.ImageRepresentationType.OME_NGFF))
example_image
```
