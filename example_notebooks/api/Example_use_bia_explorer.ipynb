```python
from bia_explorer import api
```

```python
# Get and display information about a study using its accession id
study = api.BIAStudy.get_by_accession("S-BIAD582")
study
```

```python
# View the number of images in the study.
study.images_count
```

```python
# View info of the next image in the study
next(study.get_images())
```

```python
# Calculate the number of images with OME-NGFF representations in the study
ome_ngff_images = list(study.get_images_with_representation(api.ImageRepresentationType.OME_NGFF))
len(ome_ngff_images)
```

```python
# View the metadata of the first image with an OME-NGFF representation
ome_ngff_images[0]
```

```python
# Display the first slice of the OME-NGFF image
ome_rep = [rep for rep in ome_ngff_images[0].representations if rep.type == 'ome_ngff'][0]
ome_rep.get_slice_image(api.ImageSlice(c=0,t=0,z=0))
```

```python
# Use another method to view the info of an OME-NGFF image representation
# in the study (first image in the iteration)
image_rep = next(study.get_image_representations(api.ImageRepresentationType.OME_NGFF))
image_rep
```

```python
# Display the first slice of the OME-NGFF image
image_rep.get_slice_image(api.ImageSlice(c=0,t=0,z=0))
```

```python
# Get the uri of the OME-NGFF image
print(image_rep.uri[0])
```

```python
# Calculate the number of studies in BIA API's db
len(list(api.BIAStudy.get_all()))
```

```python
# Retrieve information about collections in the BioImage Archive and print total number of collections
all_collections = list(api.BIACollection.get_all())
len(all_collections)
```

```python
# View the metadata of the first collection in the list
example_collection = all_collections[0]
example_collection
```

```python
# Get the metadata on the next study in the example_collection
next(example_collection.get_studies())
```

```python
# Print the name of all collections in BIA
', '.join([collection.name for collection in all_collections])
```

```python
# Retrieve and view the metadata of the AI collection
ai_collection = api.BIACollection.get_by_name('AI')
ai_collection
```

```python
print('Number of studies in this collection =', len(ai_collection.study_uuids))
```

```python
# View the next study in the AI collection
next(ai_collection.get_studies())
```

```python
# Use the study_uuid to select which study to retrieve
ai_study = api.ApiClient.get_study_by_uuid(ai_collection.study_uuids[2])
ai_study
```

```python
# Rertieve and view the metadata of the next image with an OME-NGFF representation
ome_ngff_image = next(ai_study.get_images_with_representation(api.ImageRepresentationType.OME_NGFF))
ome_ngff_image
```

```python
# Display the second slice of the OME-NGFF image
ome_rep = [rep for rep in ome_ngff_image.representations if rep.type == 'ome_ngff'][0]
ome_rep.get_slice_image(api.ImageSlice(c=0,t=0,z=1))
```

```python
# Display the first slice of an OME-NGFF image in the study using
# another method to get the OME-NGFF image representation
image_representation = next(ai_study.get_image_representations(
    api.ImageRepresentationType.OME_NGFF
))
image_representation.get_slice_image(api.ImageSlice(c=0,t=0,z=0))
```

```python
print(image_representation.uri[0])
```

```python
# Display the image metadata that the image representation belongs
image = [image for image in ai_study.get_images() if image_representation in image.representations][0]
image
```

```python
# Get the OME-NGFF image as a dask array
dask_array = image_representation.ome_ngff_to_dask_array()
dask_array
```

```python
from itkwidgets import view
```

```python
# Get a study by its accession id, select an OME-NGFF image of it
study_itk = api.BIAStudy.get_by_accession('S-BIAD634')
image_representation_itk = list(study_itk.get_image_representations(
    api.ImageRepresentationType.OME_NGFF
))[1]
# Read the image as a dask array and select a slice
dask_array_itk = image_representation_itk.ome_ngff_to_dask_array()
selected_slice = dask_array_itk[0,0,0,:,:].compute()
```

```python
# View using ITKviewer in the notebook
viewer = view(selected_slice)
viewer
```
